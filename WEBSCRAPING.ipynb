```python
data = {
    "Player":      [],
    "From":        [],
    "To":          [],
    "Pos":         [],
    "Ht":          [],
    "Wt":          [],
    "Birth_Date":  [],
    "Colleges":    []
}

for row in tablebody.getchildren()[0:]:
    data["Player"].append(row.getchildren()[0].text_content())
    data["From"].append(row.getchildren()[1].text_content())
    data["To"].append(row.getchildren()[2].text_content())
    data["Pos"].append(row.getchildren()[3].text_content())
    data["Ht"].append(row.getchildren()[4].text_content())
    data["Wt"].append(row.getchildren()[5].text_content())
    data["Birth_Date"].append(row.getchildren()[6].text_content())
    data["Colleges"].append(row.getchildren()[7].text_content())
    
```

```python
df=pd.DataFrame(data)
df
```

```python
df.info()
```

```python
conn   = sqlite3.connect(":memory:")
```

```python
cursor = conn.cursor()
```

```python
cursor.execute("""
    CREATE TABLE PLAYERS (
        NAME  TEXT,
        START_YEAR   NUMBER,
        END_YEAR     NUMBER,
        POSITION     TEXT,
        HEIGHT       TEXT,
        WEIGHT       TEXT,
        BIRTH_DATE   NUMBER,
        COLLEGES     TEXT
    )
""")
```

```python
df.head().values
```

```python
cursor.executemany("INSERT INTO PLAYERS VALUES (?,?,?,?,?,?,?,?)", df.values)
```

```python
conn.commit()
```
